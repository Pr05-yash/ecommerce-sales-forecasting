```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Load data
sales_data = pd.read_csv('../data/sales_data.csv')
sales_data['Date'] = pd.to_datetime(sales_data['Date'])
sales_data.set_index('Date', inplace=True)
sales_data.index = pd.DatetimeIndex(sales_data.index, freq='D')

# Plot sales
sales_data.plot(title='Daily Sales', figsize=(10, 4))
plt.grid(True)
plt.show()

# Build ARIMA model
model = ARIMA(sales_data['Sales'], order=(5, 1, 0))
model_fit = model.fit()

# Forecast next 30 days
forecast = model_fit.forecast(steps=30)

# Plot forecast
forecast.plot(title='30-Day Sales Forecast', figsize=(10, 4))
plt.grid(True)
plt.xlabel("Date")
plt.ylabel("Forecasted Sales")
plt.show()

forecast = model_fit.forecast(steps=30)

# Plot forecast
forecast.plot(title='30-Day Sales Forecast', figsize=(10, 4))
plt.grid(True)
plt.savefig('../plots/sales_forecast.png')  # Ye line last me honi chahiye
plt.show()
```
