```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

# Load data
customer_data = pd.read_csv('../data/customer_data.csv')

# Clustering karna (Segment column create hoga)
X = customer_data[['AmountSpent', 'Frequency', 'Recency']]
kmeans = KMeans(n_clusters=4, random_state=42)
customer_data['Segment'] = kmeans.fit_predict(X)

# Ab plot karna (Segment column ke saath)
sns.pairplot(customer_data, hue='Segment', vars=['AmountSpent', 'Frequency', 'Recency'])
plt.suptitle("Customer Segmentation Clusters", y=1.02)

# Plot save karna
plt.savefig('../plots/customer_segmentation.png')

plt.show()
```
